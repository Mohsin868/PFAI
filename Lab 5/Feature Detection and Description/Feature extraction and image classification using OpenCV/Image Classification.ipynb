### Step 2: Importing Necessary Libraries

```python
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Converting the image to gray
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
plt.imshow(gray)
```
