## <b>Bilateral Filtering<b>

```python
import cv2
```

```python
# Read the image.
img = cv2.imread('taj.jpg')

# Apply bilateral filter with d = 15, 
# sigmaColor = sigmaSpace = 75.
bilateral = cv2.bilateralFilter(img, 15, 75, 75)

# Save the output.
cv2.imwrite('taj_bilateral.jpg', bilateral)
```

```python
img = cv2.imread('taj.jpg')

# Apply bilateral filter with d = 15, 
# sigmaColor = sigmaSpace = 75.
bilateral = cv2.bilateralFilter(img, 5, 5, 5)

# Save the output.
cv2.imwrite('taj_bilateral1.jpg', bilateral)
```
