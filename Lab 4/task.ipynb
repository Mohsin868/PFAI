```python
# Lab 4 - N Queens Problem (Dynamic)

def is_safe(board, row, col, n):
    # Check column
    for i in range(row):
        if board[i] == col:
            return False

    # Check left diagonal
    for i in range(row):
        if board[i] - i == col - row:
            return False

    # Check right diagonal
    for i in range(row):
        if board[i] + i == col + row:
            return False

    return True


def solve_n_queens(board, row, n):
    # Base case: all queens placed
    if row == n:
        return True

    # Try each column
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            if solve_n_queens(board, row + 1, n):
                return True
            board[row] = -1  # Backtrack

    return False


def print_board(board, n):
    print("\nSolution Board:\n")
    for i in range(n):
        for j in range(n):
            if board[i] == j:
                print("Q", end=" ")
            else:
                print(".", end=" ")
        print()


# Take dynamic input
n = int(input("Enter the value of N: "))

board = [-1] * n

if solve_n_queens(board, 0, n):
    print_board(board, n)
else:
    print("No solution exists for N =", n)
```
