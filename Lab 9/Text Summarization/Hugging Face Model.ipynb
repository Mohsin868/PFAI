### Text Summarization using HuggingFace Model

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
model_name = "t5-small"

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
text = """
Artificial Intelligence (AI) is rapidly changing the world by enabling machines to perform tasks that previously required human intelligence. 
Industries such as healthcare, finance, and transportation are benefiting from AI through improved efficiency, automation, and data-driven 
decision making. For example, AI-powered systems can detect diseases early, optimize financial strategies, and even drive autonomous vehicles. 
However, despite these advantages, concerns about job displacement, data privacy, and ethical implications remain significant challenges. 
Governments and organizations are working together to create policies that ensure responsible AI development while maximizing its benefits 
for society.
"""

input_text = "summarize: " + text
```

```python
inputs = tokenizer.encode(
    input_text,
    return_tensors="pt",
    max_length=512,
    truncation=True
)
```

```python
summary_ids = model.generate(
    inputs,
    max_length=60,
    min_length=20,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)
```

```python
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("Summary:", summary)
```
