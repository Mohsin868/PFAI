```python
Jug_A = 4
Jug_B = 3
GOAL = 2

# DFS
def dfs(state, visited, path):
    if state in visited:
        return False
    visited.add(state)

    # Check goaL
    if state[0] == GOAL or state[1] == GOAL:
        path.append(state)
        return True

    # Possible actions (rules)
    a, b = state
    actions = []

    # 1. Fill Jug A
    if a < Jug_A:
        actions.append((Jug_A, b, "Fill Jug A"))
    # 2. Fill Jug B
    if b < Jug_B:
        actions.append((a, Jug_B, "Fill Jug B"))
    # 3. Empty Jug A
    if a > 0:
        actions.append((0, b, "Empty Jug A"))
    # 4. Empty Jug B
    if b > 0:
        actions.append((a, 0, "Empty Jug B"))
    # 5. Pour A -> B
    if a > 0 and b < Jug_B:
        pour = min(a, Jug_B - b)
        actions.append((a - pour, b + pour, f"Pour {pour} from A to B"))
    # 6. Pour B -> A
    if b > 0 and a < Jug_A:
        pour = min(b, Jug_A - a)
        actions.append((a + pour, b - pour, f"Pour {pour} from B to A"))

    for new_a, new_b, rule in actions:
        print(f"Applying rule: {rule} -> New state: ({new_a}, {new_b})")
        if dfs((new_a, new_b), visited, path):
            path.append(state)
            return True

    return False

# DFS
visited = set()
path = []
dfs((0,0), visited, path)

# Print solution
print("\nSolution Path (from initial state to goal):")
for state in reversed(path):
    print(state)
```
